## Women shoes dataset

This Sequential Neural Network takes simple feautures such as, Brand, Type, Color and predicts the price

### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from numpy import array
from numpy import argmax
from keras.utils import to_categorical
import keras
from keras import backend as k
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
```

### Data Preprocessing

```python
dataset = pd.read_csv('v2.csv')
```

```python
l_e = LabelEncoder()
```

```python
x1 = dataset.iloc[:,0].values
x2 = dataset.iloc[:,1].values
x3 = dataset.iloc[:,2].values
Y = dataset.iloc[:,3].values
x1 = l_e.fit_transform(x1)
x2 = l_e.fit_transform(x2)
x3 = l_e.fit_transform(x3)
x1 = to_categorical(x1)
x2 = to_categorical(x2)
x3 = to_categorical(x3)
x1=array(x1)
x2 = array(x2)
x3 = array(x3)

k = x1.shape[1]+x2.shape[1]+x3.shape[1]

X = np.zeros((len(x1),k))
for i in range(len(x1)):
    l = []
    for j in x1[i]:
        l.append(j)
    for j in x2[i]:
        l.append(j)
    for j in x3[i]:
        l.append(j)
    X[i]=array(l)
    
```

### Creating the model

```python
model = Sequential([Dense(32, input_shape = [556,], activation = 'relu'),Dense(32, activation = 'relu'),Dense(1,activation='relu')])
```

```python
model.summary()
```

### Compile the model

```python
model.compile(Adam(),loss = 'mse' )
```

### Train the model

```python
model.fit(X,Y,batch_size = 20,shuffle=True,verbose=0,epochs = 10)
```
